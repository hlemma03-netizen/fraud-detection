```python
import pandas as pd

merged_df = pd.read_csv(
    "../data/processed/fraud_with_country.csv"
)

merged_df["signup_time"] = pd.to_datetime(
    merged_df["signup_time"]
)

merged_df["purchase_time"] = pd.to_datetime(
    merged_df["purchase_time"]
)
```

```python
merged_df["time_since_signup"] = (
    merged_df["purchase_time"]
    - merged_df["signup_time"]
).dt.total_seconds()
```

```python
merged_df["hour_of_day"] = (
    merged_df["purchase_time"].dt.hour
)
```

```python
merged_df["day_of_week"] = (
    merged_df["purchase_time"].dt.dayofweek
)
```

```python
merged_df["transaction_count"] = (
    merged_df.groupby("user_id")
    ["user_id"]
    .transform("count")
)
```

```python
merged_df["device_usage_count"] = (
    merged_df.groupby("device_id")
    ["device_id"]
    .transform("count")
)
```

```python
categorical_cols = [
    "source",
    "browser",
    "sex",
    "country"
]

df_encoded = pd.get_dummies(
    merged_df,
    columns=categorical_cols,
    drop_first=True
)
```

```python
from sklearn.preprocessing import StandardScaler

scaler = StandardScaler()

num_cols = [
    "purchase_value",
    "age",
    "time_since_signup",
    "transaction_count",
    "device_usage_count"
]

df_encoded[num_cols] = scaler.fit_transform(
    df_encoded[num_cols]
)
```
